```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
```

```python
def generate_data(n, t, f0):
    A = 5
    R = 1
    data = dict()
    H = 10
    data[0] = 10
    for i in range(n):
        H  = H + t/A*(-f0(H))
        data[t*(i+1)] = H
    return data
```

```python
def draw_data(n, t, fig, f0):
    data = generate_data(n, t, f0)
    fig.add_trace(
        go.Scatter(x = list(data.keys()), y = list(data.values()), name=f"N = {n}, T = {t}"),
        row = 1, col = 1
    )
```

#### EXERCISE 1.5

```python

N = [10, 25, 100]
T = [2.5, 1, 0.25]
R = 1
fig = make_subplots(rows=1, cols=1)
for n, t in zip(N, T):
    draw_data(n, t, fig, lambda H: H/R)

fig.show()
```

```python
N = [10, 25, 100]
T = [2.5, 1, 0.25]
c = 3
fig = make_subplots(rows=1, cols=1)
for n, t in zip(N, T):
    draw_data(n, t, fig, lambda H: c * max(0,H)**0.5)

fig.show()
```
