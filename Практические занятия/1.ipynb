```python
import math
from plotly.subplots import make_subplots
import pandas as pd
```

```python
fig = make_subplots(rows=2, cols=2)

Ts = [0.01, 0.025, 0.05, 0.1]
t = 2
diff = [0, 0, 0, 0]

for j,T in enumerate(Ts):
    y_true = [1.75 * math.exp(-2 * t * T) + 1.5 * t * T - 0.75 for t in range(int(2 / T) + 1)]
    df = {'x':[],'y':[]}
    y = [0 for _ in range(int(2 / T) + 1)]
    y[0] = 1
    df['x'].append(0)
    df['y'].append(1)
    for i in range(1, len(y)):
        y[i] = (1 - 2 * T) * y[i - 1] + 3 * (i - 1) * (T**2)
        diff[j] += (y[i] - y_true[i]) ** 2
        df['x'].append(i*T)
        df['y'].append(y[i])
    df = pd.DataFrame(df)

    fig.add_scatter(
        x=df['x'],y=df['y'],
        row=j//2+1,col=j%2+1,
        mode='markers',
        marker=dict(
            color='Black',
            size=3,
        )
    )
    fig.add_scatter(
        x=df['x'],y=y_true,
        row=j//2+1,col=j%2+1,
        line=dict(
                color='MediumPurple',
                width=2
            )
    )

fig.show()
```

```python
T = 0.25
N = int(100 / T + 1)
A = 10
R = 2
H = [0 for _ in range(N+1)]
H[0] = 4
fig = make_subplots(rows=1, cols=1)
df = {'x':[],'y':[]}
df['x'].append(0)
df['y'].append(4)
for n in range(N):
    H[n+1] = (1- T/(A*R)) * H[n] + (T/A) * 5
    df['x'].append(T*n)
    df['y'].append(H[n])
df = pd.DataFrame(df)
fig.add_scatter(
        x=df['x'],y=df['y'],
        row=1,col=1,
        mode='markers', 
        marker=dict(
            color='Black',
            size=2,
        )
    )

```
